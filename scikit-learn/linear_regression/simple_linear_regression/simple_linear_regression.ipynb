# Simple Linear Regression

## Import Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error
```

## Load Dataset

```python
df = pd.read_csv('Salary_dataset.csv')
df.head()
```

```python
df = df[["YearsExperience", "Salary"]]
df.head()
```

```python
# total number of rows and columns
len(df)
```

```python
# check for missing data
df.isnull().sum()
```

## Visualize the dataset

```python
experience = df[["YearsExperience"]]
salary = df[["Salary"]]
plt.scatter(experience, salary, color='red', marker='o')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Split the dataset

```python
X_train, X_test, y_train, y_test = train_test_split(experience, salary, test_size=0.2, random_state=42)
```

```python
X_train.head(), X_test.head(), y_train.head(), y_test.head()
```

```python
# total number of rows in the training set
len(X_train), len(y_train)
```

```python
# total number of rows in the test set
len(X_test), len(y_test)
```

## Create Linear Regression Model

```python
model = linear_model.LinearRegression()
model
```

## Train The Model

```python
model.fit(X_train, y_train)
```

## Make Predictions and Evaluate the Model

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# calculate the root mean squared error
rmse = root_mean_squared_error(y_test, y_pred)
print(f"Root Mean Squared Error: {rmse}")
```

## Visualize the Results

```python
# plotting the training data
plt.scatter(X_train, y_train, color='blue', label='Training Data')

# plotting the test data
plt.scatter(X_test, y_test, color='green', label='Test Data')

# plotting the regression line
plt.plot(experience, model.predict(experience), color='red', label='Regression Line')

plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
