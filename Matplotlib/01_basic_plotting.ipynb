## Matplotlib Basic Plotting
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

## Importing Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Line Graph
Line graphs are used to track changes over short and long periods of time. When smaller changes exist, line graphs are better to use than bar graphs. Line graphs can also be used to compare changes over the same period of time for more than one group. In a line graph, data points are plotted on a graph and then connected by a line

```python
x = [10, 20]
y = [30, 40]

plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='.')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
np.random.seed(0)
x = np.random.rand(3)
y = np.random.rand(3)
print(x, y)


plt.plot(x, y, marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_sin, label='Sine Wave', color='r', marker='o',)
plt.plot(x, y_cos, label='Cosine Wave', color='b', marker='.')

plt.title('Sine and Cosine Waves')
plt.xlabel('X values (radians)')
plt.ylabel('Y values')
plt.grid(True)
plt.legend()
plt.show()
```

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
temperature = [22, 24, 20, 23, 25, 27, 26]

plt.figure(figsize=(10, 6))
plt.plot(days, temperature, label='Temperature', marker='o')

plt.title('Temperature of this Week')
plt.xlabel('Day')
plt.ylabel('Temperature (celsius)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

## Bar Graph
In a bar graph, data is represented by rectangular bars where the length of the bar is proportional to the quantity of the data. A bar chart describes the comparisons between the discrete categories.

**Basic Vertical Bar Graph**

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

plt.figure(figsize=(8, 6))
plt.bar(products, sales, color='b')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales of Different Products')
plt.show()
```

**Stacked Bar Graph**

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
online_sales = [20, 15, 30, 25]
offline_sales = [5, 10, 5, 10]

plt.figure(figsize=(10, 6))
plt.bar(products, online_sales, label='Online Sales')
plt.bar(products, offline_sales, label='Offline Sales', bottom=online_sales)

plt.xlabel('Products')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

**Grouped Bar Graph**

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
online_sales = [20, 15, 30, 25]
offline_sales = [5, 10, 5, 10]

bar_width = 0.35
bar_position1 = np.arange(len(products))
bar_position2 = bar_position1 + bar_width

plt.figure(figsize=(10, 6))
plt.bar(bar_position1, online_sales, bar_width, label='Online Sales', color='skyblue')
plt.bar(bar_position2, offline_sales, bar_width, label='Offline Sales', color='orange')

plt.xlabel('Products')
plt.ylabel('Sales')
plt.xticks(np.arange(len(products)), products)
plt.legend()
plt.show()
```

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
region1_sales = [23, 17, 35, 29]
region2_sales = [15, 22, 30, 21]

x = np.arange(len(products))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x + width/2, region1_sales, width, label='Region 1')
plt.bar(x - width/2, region2_sales, width, label='Region 2')
plt.title('Sales Comparison by Region')
plt.xticks(x, products)
plt.legend()
plt.show()
```

**Horizontal Bar Graph**

```python
subjects = ['Math', 'Science', 'English', 'History']
grades = [85, 90, 78, 88]

plt.figure(figsize=(10, 6))
plt.barh(subjects, grades, color=['green', 'yellow', 'orange', 'blue'])
plt.xlabel('Grades')
plt.ylabel('Subjects')
plt.title('Average Grades by Subject')
plt.show()
```
